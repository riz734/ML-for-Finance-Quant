```python
from pytube import YouTube
import numpy as np
import datetime
import pandas as pd
import whisper
import datetime

```

```python
#three part series
part1 = YouTube("https://www.youtube.com/watch?v=386p68_lDHA")
part2 = YouTube("https://www.youtube.com/watch?v=wvzyDg40-yw")
part3 = YouTube("https://www.youtube.com/watch?v=8-fugWMBwCg")
```

```python
#capturing only available audio streams
streams1 = part1.streams.filter(only_audio=True)
streams2 = part2.streams.filter(only_audio=True)
streams3 = part3.streams.filter(only_audio=True)
```

```python
#selects the first available audio streams and downloads them
streams1 = streams1.first()
streams1.download(filename='part1_Coffee_LP.mp4')

streams2 = streams2.first()
streams2.download(filename='part2_Coffee_LP.mp4')

streams3 = streams3.first()
streams3.download(filename='part3_Coffee_LP.mp4')
```

```python
#sets up a whisper model to transcribe audio
model = whisper.load_model('base')
```

```python
# do the transcriptions, timestamps how long it takes

t1 = datetime.datetime.now()
print(f"started at {t1}")

transcribed_p1 = model.transcribe("part1_Coffee_LP.mp4")
transcribed_p2 = model.transcribe("part2_Coffee_LP.mp4")
transcribed_p3 = model.transcribe("part3_Coffee_LP.mp4")

t2 = datetime.datetime.now()

print(f"ended at {t2}")
print(f"time elapsed: {t2 - t1}")
```

```python
complete_series = transcribed_p1["text"] + transcribed_p2["text"] + transcribed_p3["text"]
complete_series
```

Using nltk to clean up text

```python
#!pip install nltk
import nltk

from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize

#splitting up the transcript into smaller pieces (by sentences), tokens
transcript_sent=sent_tokenize(complete_series)
transcript_sent
```

```python
#nltk.download('stopwords')
from nltk.corpus import stopwords
#grabs stop words into a list
#stop words add noise, words like "a", "that", "am",etc.
stop_words=set(stopwords.words("english"))
filtered = []
for i in transcript_sent:
    if i not in stop_words:
        #print(i)
        filtered.append(i)
filtered
```

```python
stop_words
```

```python
filtered
```

```python
# reducing words to root, i.e. filtering > filter
from nltk.stem import PorterStemmer
from nltk.tokenize import sent_tokenize, word_tokenize

ps = PorterStemmer()
stemmed_words=[]
for w in filtered:
    stemmed_words.append(ps.stem(w))

#print("Filtered Sentence:",filtered)
print("Stemmed Sentence:",stemmed_words)
```

```python
model_df = pd.DataFrame(stemmed_words)
model_df = model_df.rename(columns={0:"text"})
```

```python
# calculating and adding sentiment scores to dataframe
from nltk.sentiment.vader import SentimentIntensityAnalyzer

def sentiment_scores(df):
    
    senti_lyze = SentimentIntensityAnalyzer()
    df.reset_index(inplace=True, drop=True)
    df[['neg', 'neu', 'pos', 'compound']] = df['text'].apply(senti_lyze.polarity_scores).apply(pd.Series)
    
    return df


#neg for negative sentiment
#neu for neutral sentiment
#pos for positive sentiment
#compound for an overall score that combines negative, positive, and neutral sentiments into a single score.

senti_df = sentiment_scores(model_df)
```

```python
import plotly.express as px
import seaborn as sns
px.scatter(senti_df[senti_df["neg"]!=0], x='compound', y='text')
```

```python
with open("text_output_neg.txt", "w") as text_file:
    neg_senti_df = senti_df[senti_df["neg"]!=0]
    for i in neg_senti_df["text"]:
        text_file.write(i)
  
```
